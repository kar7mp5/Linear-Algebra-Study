### Lagrange Interpolation formula

$$
f_i(x) = \prod^n__{k=0 k \neq i}{\frac{x - c_k}{c_i - c_k}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

points = np.array([[1, 8], [2, 5], [3, -4]])

def func(x) -> np.array:
  return -3*x**2 + 6*x + 5

X = np.arange(-20, 21)
Y = func(X)

plt.figure(figsize=(14, 6))
plt.scatter(points[:, 0], points[:, 1])
plt.plot(X, Y)
```
